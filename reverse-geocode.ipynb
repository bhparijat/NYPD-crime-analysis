```python
import requests
import pandas as pd 
import time
from multiprocessing import Pool
```

```python
data = pd.read_csv('data/NYPD_Arrests_Data__Historic_.csv')
```

```python
lat_long = data[['Latitude','Longitude']]
```

```python


base_url = "http://34.83.157.143:4000/v1/reverse?"
query_params = "&sources=oa"
```

```python
df = pd.DataFrame(data={'postalcode':[-1],'point_no':[-1]})
df = df.append({'postalcode':1,'point_no':0},ignore_index=True) 
df
```

```python
start = time.time()
count = 0 
#base_url = "http://photon.komoot.de/reverse?"

df = pd.DataFrame(data={'postalcode':[-1],'point_no':[-1]})

not_present = []
for i,row in lat_long.iloc[:,:].iterrows():
    url = base_url + "point.lon=" + str(row[1]) + "&point.lat=" + str(row[0]) + query_params
    res = requests.request('GET',url).json()
    if i%100 == 0:
        taken = (time.time()-start)
        print("{} done after {}".format(i,taken))
        
    properties = res['features'][0]['properties']
    
    if len(res['features'])>1:
        count+=1
    #break
    if 'postalcode' not in properties:
        not_present.append(i)
        print("postcode not present for i = {}".format(i))
        continue
        
    postal_code =  properties['postalcode']
    df = df.append({'postalcode':postal_code,'point_no':i},ignore_index=True)    
    if i%1000 == 0:
        df.to_csv(('reverse'+str(i))+'.csv',index=False)
     
```
